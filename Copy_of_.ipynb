```python

df = pd.read_csv("indian_startups_fundingDADS.csv")

df.head()
df.shape
df.isnull()
df.isnull().sum()
df.duplicated().sum()
# Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

# Print cleaned column names in alphabetical order
print(sorted(df.columns))

stats = {
    "count": df["founded_year"].dropna().count(),
    "mean": df["founded_year"].dropna().mean(),
    "min": df["founded_year"].dropna().min(),
    "max": df["founded_year"].dropna().max(),
    "median": df["founded_year"].dropna().median(),
    "std": df["founded_year"].dropna().std()
}

print(stats)

city_map = {
    # Ahmedabad
    "AHMEDABAD": "Ahmedabad",
    "ahmedabad": "Ahmedabad",

    # Bengaluru
    "BANGALURU": "Bengaluru",
    "Bangalore": "Bengaluru",
    "bangalore": "Bengaluru",
    "bengaluru": "Bengaluru",
    "BENGALURU":"Bengaluru",

    "BLR": "Bengaluru",

    # Mumbai
    "BOM": "Mumbai",
    "Bombay": "Mumbai",
    "MUMBAI": "Mumbai",
    "mumbai": "Mumbai",

    # Chennai
    "CHENNAI": "Chennai",
    "chennai": "Chennai",
    "Madras": "Chennai",

    # Delhi
    "DELHI": "Delhi",
    "delhi": "Delhi",
    "NEW DELHI": "Delhi",
    "New Delhi": "Delhi",

    # Gurugram
    "GURGAON": "Gurugram",
    "Gurgaon": "Gurugram",
    "gurgaon": "Gurugram",
    "GGN": "Gurugram",
    'GURUGRAM':"Gurugram",

    # Hyderabad
    "HYD":"Hyderabad",
    "HYDERABAD": "Hyderabad",
    "hyderabad": "Hyderabad",

    # Kolkata
    "KOLKATA": "Kolkata",
    "Calcutta": "Kolkata",

    # Noida
    "NOIDA": "Noida",

    # Pune
    "PUNE": "Pune",
    "pune": "Pune"
}

df["city"] = df["city"].replace(city_map)
print(sorted(df["city"].unique()))
industry_map = {

    # AI/ML
    "AGRITECH": "Agritech",
    "agritech": "Agritech",
    "Agri Tech": "Agritech",
    "Agri-Tech": "Agritech",
    'AgriTech':"Agritech",

    "AI ML": "AI/ML",
    "AI-ML": "AI/ML",
    "AI/ML": "AI/ML",
    "AIML": "AI/ML",

    # Beauty
    "BEAUTYTECH": "BeautyTech",
    "Beauty-Tech": "BeautyTech",

    # Ecommerce
    "E-Commerce": "Ecommerce",
    "E-commerce": "Ecommerce",
    "ECOMMERCE": "Ecommerce",
    "ecommerce": "Ecommerce",

    # EdTech
    "EDTECH": "EdTech",
    "Ed-Tech": "EdTech",
    "Edtech": "EdTech",

    # EV
    "EV": "EV",
    "Electric Vehicles": "EV",
    "Electric-Vehicles": "EV",
    "ev": "EV",

    # FinTech
    "FINTECH": "FinTech",
    "Fin-Tech": "FinTech",
    "Fintech": "FinTech",
    "fintech": "FinTech",
    "FT": "FinTech",

    # FoodTech
    "FOOD-TECH": "FoodTech",
    "Food Tech": "FoodTech",
    "Food-Tech": "FoodTech",
    "foodtech": "FoodTech",

    # Gaming
    "GAMING": "Gaming",
    "Games": "Gaming",
    "gaming": "Gaming",

    # HealthTech
    "HEALTHTECH": "HealthTech",
    "Health Tech": "HealthTech",
    "Health-Tech": "HealthTech",
    "healthtech": "HealthTech",

    # Logistics
    "LOGISTICS": "Logistics",
    "logistics": "Logistics",

    # PropTech
    "PROPTECH": "PropTech",
    "Prop-Tech": "PropTech",

    # SaaS
    "SAAS": "SaaS",
    "SaaS": "SaaS",
    "Saas": "SaaS",
    "saas": "SaaS",

    # TravelTech
    "TRAVELTECH": "TravelTech",
    "Travel-Tech": "TravelTech"
}
df["industry"] = df["industry"].str.strip().replace(industry_map)
print(sorted(df["industry"].unique()))
print(df["industry"].value_counts())
# Convert amount column to numeric
df["amount"] = pd.to_numeric(
    df["amount"].replace("Undisclosed", np.nan),
    errors="coerce"
)

# (a) Datatype
print("Data Type:")
print(df["amount"].dtype)

# (b) Number of NaN values
print("\nNaN Values:")
print(df["amount"].isna().sum())

# (c) Mean funding amount
print("\nMean Funding Amount:")
print(df["amount"].mean())

# (d) Minimum and Maximum funding amount
print("\nMinimum Funding Amount:")
print(df["amount"].min())

print("\nMaximum Funding Amount:")
print(df["amount"].max())
# Convert funding_date to datetime
df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce"
)

# (a) Count deals in each year
print(df["funding_date"].dt.year.value_counts().sort_index())

# (b) Year with highest funding deals
print("\nYear with highest funding deals:")
print(df["funding_date"].dt.year.value_counts().idxmax())

df["funding_stage"] = df["funding_stage"].str.strip().str.title()
stage_map = {

    # Bridge
    "Bridge": "Bridge",

    # IPO
    "Ipo": "IPO",

    # Seed
    "Pre Seed": "Pre-Seed",
    "Pre-Seed": "Pre-Seed",
    "Preseed": "Pre-Seed",

    "Seed": "Seed",

    # Series A
    "Pre Series A": "Pre-Series A",
    "Pre-Series A": "Pre-Series A",
    "Pre-A": "Pre-Series A",

    "Series A": "Series A",
    "Series-A": "Series A",
    'Series_A':"Series A",

    # Series B
    "Series B": "Series B",
    "Series-B": "Series B",
    "Series_B":"Series B",

    # Series C
    "Series C": "Series C",
    "Series-C": "Series C",

    # Series D
    "Series D": "Series D",
    "Series-D": "Series D",

    # Series E
    "Series E": "Series E",
    "Series-E": "Series E"
}
df["funding_stage"] = df["funding_stage"].replace(stage_map)
print(sorted(df["funding_stage"].unique()))
print(df["funding_stage"].value_counts())
result = df[(df["city"] == "Bengaluru") & (df["amount"] > 100)]

result = result.sort_values(by="amount", ascending=False)

print(result[["startup_name", "amount"]])
result = df[df["industry"].isin(["FinTech", "EdTech", "HealthTech"])]

print(len(result))

result = df[df["startup_name"].str.contains("pay", case=False, na=False)]

print(result["startup_name"].unique())

df["funding_efficiency"] = df["amount"] / (2024 - df["founded_year"])

result = df.sort_values(by="funding_efficiency", ascending=False)

print(result[["startup_name", "industry", "amount", "founded_year", "funding_efficiency"]].head(10))

df["funding_tier"] = df["amount"].apply(
    lambda x:
    "Unicorn" if x >= 1000 else
    "Mega" if x >= 500 else
    "Large" if x >= 100 else
    "Mid" if x >= 10 else
    "Small / Unknown"
)

print(df["funding_tier"].value_counts())
df["era"] = pd.cut(
    df["founded_year"],
    bins=[0, 2009, 2017, float("inf")],
    labels=["Early", "Growth", "Recent"]
)

print(df["era"].value_counts())

df["investor_count"] = df["investors"].apply(
    lambda x: 0 if pd.isna(x) else len(x.split(", "))
)

print(df["investor_count"].value_counts())

result = (
    df.groupby("industry")["amount"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

print(result)

result = (
    df.groupby("city")
      .agg({
          "startup_name": "count",
          "amount": ["sum", "mean"]
      })
)

result.columns = ["startup_count", "total_funding", "average_funding"]

print(result.sort_values(by="startup_count", ascending=False).head(10))


result = (
    df.groupby("industry")
      .agg(
          count=("startup_name", "count"),
          average_funding=("amount", "mean")
      )
)

result = result.query("count >= 15")

print(result.sort_values("average_funding", ascending=False).head(1))

top5 = df["industry"].value_counts().head(5).index

print(pd.crosstab(df["industry"], df["funding_tier"]).loc[top5])

idx = df.groupby("industry")["amount"].idxmax()

print(df.loc[idx, ["startup_name", "industry", "city", "amount"]])

result = df[
    (df["founded_year"] >= 2019) &
    (df["amount"] >= 100) &
    (df["industry"].isin(["FinTech", "EdTech", "HealthTech"]))
]

print(
    result.sort_values("amount", ascending=False)[
        ["startup_name", "industry", "city", "founded_year", "amount"]
    ]
)

result = (
    df.groupby("city")
      .agg(
          count=("startup_name", "count"),
          total_funding=("amount", "sum")
      )
)

result = result.query("count >= 20")

print(result.sort_values("total_funding", ascending=False))

result = (
    df.groupby(["city", "industry"])
      .size()
      .reset_index(name="startup_count")
      .sort_values("startup_count", ascending=False)
      .head(10)
)

print(result)
```
